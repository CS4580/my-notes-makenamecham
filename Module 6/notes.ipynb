### Task: Understand Museum Attendance

```python
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print('setup complete')
```

```python
# Load data 
data_path = '../data/museum_visitors.csv'
museum_df = pd.read_csv(data_path, index_col='Date', parse_dates=True)


museum_df.head()
```

The Firehouse museum claims they ran an event in 2014 that brought a big number of visitors, and they should get an extra budget to run a similar event this year.

Task: Show the museum board how the event compared to regular traffic at each museum.
- Create a `lineplot chart` that shows how the number of visitors change to each museum change over time 
- Provide a title for the plot

```python
plt.figure(figsize=(16,6))
plt.title('Museum attendance in 2014')

sns.lineplot(data=museum_df['Avila Adobe'], label = 'Avila Adobe')
sns.lineplot(data=museum_df['Firehouse Museum'], label = 'Firehouse Museum')
sns.lineplot(data=museum_df['Chinese American Museum'], label = 'Chinese American Museum')
sns.lineplot(data=museum_df['America Tropical Interpretive Center'], label = 'America Tropical Interpretive Center')

plt.xlabel('Date')
plt.ylabel('Number of Visitors')
```

### Assess Seasonality
When meeting with the employees of the `Avila Adobe`, you hear that one major pain point is that the number of the visitors varies greatly with the season.

Part A:
- Create a `line chart` that shows how the number of visitors to the `Avila Adobe` has evolved over time
- Replace `inf` with `nan`, and then fill them in with 0 

```python
import numpy as np
# Replace infinite vales with NaN
museum_df = museum_df.replace([np.inf, -np.inf], np.nan)
# Now we can use the fillna() method to replace NaN with zeros
museum_df = museum_df.fillna(0)
```

```python
# Plot the data
plt.figure(figsize=(16,6))
sns.lineplot(data=museum_df['Avila Adobe'], label = 'Avila Adobe')
plt.title('monthly visitors to Avila Adobe')
plt.xlabel('Date')
```

### Bar Plots
Dataset: US Department of Transportation that tracks flight delays

```python
# Load DataSet
file_path = '../data/flight_delays.csv'
flight_delay_df = pd.read_csv(file_path, index_col= 'Month')

flight_delay_df.head() 
```

```python
# Use a bar plot
plt.figure(figsize=(16,6))
plt.title('Average Arrival Delay for Spirit Airlines, by month')

sns.barplot(x=flight_delay_df.index, y=flight_delay_df['NK'])
plt.ylabel('Arrival Delay in min')
```

# Heat Maps

```python
# Plot heatmap for all data
plt.figure(figsize=(16,6))
plt.title('Avg Arrival Delay for Each Airline by Month')

sns.heatmap(data=flight_delay_df, annot=True)
plt.xlabel('Airline')
sns.heatmap(data=flight_delay_df, cmap='coolwarm')
```
